```python
def get_values(array):
    prom = array.mean()
    sum = 0
    for i in array:
        sum += (i-prom)**2
    ret = math.sqrt(sum)
    return prom, ret
```

```python
def get_values_list(array, sum):
    prom = sum/len(array)
    sum = 0
    for i in array:
        sum += (i-prom)**2
    ret = math.sqrt(sum)
    return prom, ret
```

```python
def get_m(array):
    array_x = array['I'].to_list()
    array_y = array['R'].to_list()
    m = (array_y[-1] - array_y[0])/(array_x[-1] - array_x[0])
    return m

def get_tasa_de_crecimiento_cada_10_iteraciones(array):
    array_celdas_vivas = array['N'].to_list()
    array_de_incrementos = []
    for i in range(10, len(array_celdas_vivas), 10):
        array_de_incrementos.append(array_celdas_vivas[i] - array_celdas_vivas[i-10])
    return np.array(array_de_incrementos).mean()
```
