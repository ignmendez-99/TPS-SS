```python
def get_values(array):
    prom = array.mean()
    sum = 0
    for i in array:
        sum += (i-prom)**2
    ret = math.sqrt(sum)
    return prom, ret
```

```python
def get_values_list(array, sum):
    prom = sum/len(array)
    sum = 0
    for i in array:
        sum += (i-prom)**2
    ret = math.sqrt(sum)
    return prom, ret
```
